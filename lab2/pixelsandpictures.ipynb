# Gradients
Montfort describes the following technique for creating a *linear* gradient, or one where the color changes smoothly across an image.

````python
from PIL import Image
rectangle = Image.new('RGB', (150, 100))
for x in range(150):
    for y in range(100):
        rectangle.putpixel((x, y), (x, x, x))
        rectangle.save('gradient.png')
rectangle
````

````python
#try2
verimage = Image.open('example.png')
verimage

def thumbnail(img):
    mode = img.mode
    size_x = int(round(img.size[0] / 3))
    size_y = int(round(img.size[1] / 3))
    new_size = (size_x, size_y)
    img = Image.new(mode, new_size)
    return img

thumbnail(verimage)
````
